```python
!apt-get update
!apt-get install -y libopencv-dev python3-opencv
!pip install numpy opencv-python
import cv2
import numpy as np
from google.colab import files

# Upload files
uploaded = files.upload()

# Function to extract features from an image
def extract_features(image_path, target_size=(128, 128)):
    # Read the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise FileNotFoundError(f"Unable to load image: {image_path}")

    # Resize the image to the target size
    img_resized = cv2.resize(img, target_size)

    # Flatten the resized image into a 1D feature vector
    features = img_resized.flatten().astype(np.float64)
    return features

# Function to compute LTLp distance
def compute_LTLp_distance(features1, features2, p):
    if len(features1) != len(features2):
        raise ValueError("Feature vectors must have the same length.")
    return np.sum(np.abs(features1 - features2) ** p) ** (1 / p)

# Main logic
image1 = "face1.jpg"  # Ensure you upload this file
image2 = "face2.jpg"  # Ensure you upload this file

# Extract features from both images (resizing to 128x128 by default)
features1 = extract_features(image1)
features2 = extract_features(image2)

# Compute the Lp distance
p = 2  # Euclidean distance
try:
    distance = compute_LTLp_distance(features1, features2, p)
    print(f"LTLp Distance between images (p={p}): {distance}")
except Exception as e:
    print(f"Error: {e}")
from google.colab import files

# Prompt the user to upload files
uploaded = files.upload()  # Opens a file browser for upload
```
